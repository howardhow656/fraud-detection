```python
import pandas as pd
import numpy as np
import os
import warnings
import matplotlib as mp
import seaborn as sns

warnings.filterwarnings(action='ignore')
```

```python
for dirname, _, filenames in os.walk('./data/'):
    for filename in filenames:
        path = os.path.join(dirname, filename) 
        print(path.ljust(100), os.path.getsize(path) // 1000000, 'MB')
```

```python
train_trans_df = pd.read_csv('./data/train_transaction.csv ')
train_iden_df = pd.read_csv('./data/train_identity.csv')
train_data = pd.merge(train_trans_df,train_iden_df,on='TransactionID',how='left')
train_data
```

```python
test_trans_df = pd.read_csv('./data/test_transaction.csv')
test_ident_df = pd.read_csv('./data/test_identity.csv')
test_data = pd.merge(test_trans_df,test_ident_df,on='TransactionID',how='left')
test_data
```

```python
missin_row = []
train_missin = train_data.isnull().any(axis=0)
for i in range(len(train_missin)):
    if train_missin[i] == True:
        missin_row.append(train_missin.index[i])
missin_row
```
